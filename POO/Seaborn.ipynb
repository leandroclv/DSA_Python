```python
#Valores randmicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

#Variveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

#DataFrame
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

#Cria os dados para a varivel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'No Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
#Style
sea.set(style = 'ticks')

#Lmplot
sea.lmplot(x = 'altura', y = 'peso', data = dados, hue = 'flag_fumante', palette = ['tab:blue', 'tab:orange'], height = 7)

#Labels e ttulos
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relao entre altura e peso de fumantes e no fumentes')

#Remove as bordas
sea.despine()

#Show
plt.show()
```
